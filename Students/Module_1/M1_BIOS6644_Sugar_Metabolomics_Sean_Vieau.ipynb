```python
##################################################################################################################
#
# YouDo:
#  1) Remove any rows with na values
# 
#  2) Split the data into "test" and "train" bits with ~30% in the test set.
#       The "Label" column is the "y" data.  All other columns are the "X" data
#       Give them these names:
#       X_train, X_test, y_train, y_test

#######################################  BEGIN STUDENT CODE  #####################################################

# 1) Remove any rows with na values
df3.dropna(axis=0, inplace=True)
df3


#2) Spliting df3 into a test set of 30% and training set of 70%

#slicing out the label column and saving it to label. Then deleting that column from df3
label = df3[["Label"]]
del df3["Label"]

#splitting df3 into X_train and X_test
X_train = df3.sample(frac=0.7)
X_test = df3.drop(X_train.index)

#splitting label into y_train and y_test
y_train = label.sample(frac=0.7)
y_test = label.drop(y_train.index)


#######################################   END STUDENT CODE   #####################################################
```

```python
print(X_train.shape, X_test.shape)

print(X_train)
print(X_test)
```

```python
print(y_train.shape, y_test.shape)

print(y_train)
print(y_test)
```

```python
#Build a model to classify the Label based on the other columns

from sklearn.ensemble import RandomForestClassifier


clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=8675309)
clf.fit(X_train, y_train.values.ravel()) #I had to change this to y_train.values.ravel() 
#I guess since I made y_train a data frame and it was just asking for a single column of label,
# without the index column?
```

```python
# Test the model
predictions = clf.predict(X_test)
reality = y_test

print(predictions)
print(reality)
```

In this code we will calculate the accuracy score of our model on the test and training data sets.

We will also plot the confusion matrices to take a closer look and assess false positives (top right) and false negatives (bottom left):

```python
# YouDo:
#  1) Evaluate the "goodness" of the predictions and write a paragraph describing the results as a markdown.
#
#######################################  BEGIN STUDENT CODE  #####################################################
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.metrics import confusion_matrix
import seaborn as sns

#Calculating the accuracy score of the testing set
test_accuracy = accuracy_score(reality, predictions)
print("Test accuracy:", test_accuracy)

#Predict the labels of the training set
y_train_pred = clf.predict(X_train)

#calculating the accuracy score of the training set
train_accuracy = accuracy_score(y_train, y_train_pred)
print("Train accuracy:", train_accuracy)

#Creating a confusion matrix for the testing set
test_confusion_matrix = confusion_matrix(reality, predictions)

# Plot the confusion matrix for the testing set
plt.figure()
sns.heatmap(test_confusion_matrix, annot=True, cmap='Blues')
plt.title('Confusion matrix for the testing set')
plt.show()

#Creating the confusion matrix for the training set
train_confusion_matrix = confusion_matrix(y_train, y_train_pred)

# Plot the confusion matrix for the training set
plt.figure()
sns.heatmap(train_confusion_matrix, annot=True, cmap='Blues')
plt.title('Confusion matrix for the training set')
plt.show()

#######################################   END STUDENT CODE   #####################################################
```

This code tells us sensitivity, specificity, PPV, and NPV of our model, which we can use to assess algorithm performance:

```python
### Sensitivity
#Recall_score is sensitivity, AKA your true positive rate
recall = recall_score(reality, predictions)
print("Sensitivity", recall)

### Specificity:
#Recall (sensitivity) for the negative class is equivalent to specificity.
#You can calculate it by inverting both the true and predicted boolean values:
specificity = recall_score(np.logical_not(reality), np.logical_not(predictions))
print("Specificity:", round(specificity, 2))

### Positive Predictive Value
#Precision_score is positive predictive value (PPV)
precision = precision_score(reality, predictions)
print("Positive Predictive Value:", precision)

### Negative Predictive Value (NPV)
#I guess we have to calculate this manually but there were 0 false negs so it's going to equal 1
# 16 / (16 + 0) = 1
print("Negative Predictive Value:", 1.0)
```

# Analysis

### Summary - Methods
A random forest classifier is a supervised machine learning algorithm used for classification.

In this notebook we took a sugar metabolomics data set investigating Chronic Kidney Disease (CKD) and cleaned it, and applied a random forest classifier to predict whether each participant had CKD.

We split the data set into a training (70%) and test (30%) set. The training set is used to fit the model, and the resulting equation is then applied to the test set. Since we know the values of 'Label' (0 = no CKD, 1 = CKD, presumably), we can compare the accuracy of the model.

### Summary - Accuracy Score

We used Sci-kit learn to compute the accuracy score of our test and training sets, with a test accuracy of 0.91 and a train accuracy of 0.96. This is pretty good! A good fit would have high test and train accuracy scores, which would indicate that the model can generalize well to new data (i.e. the test set), without overfitting or underfitting. A low test accuracy score and high train accuracy score would suggest overfitting, meaning we made the model too complex and it was trained really well on all of the features of the training set, but so much so that we couldn't generalize the model to new data. A low test and training score would suggest underfitting, meaning the model was too simple and didn't capture enough of the features in the training set, and thus couldn't extrapolate to new data.

### Summary - Confusion Matrix

Examining the heat maps gives us a closer look at the types of predictions the algorithm is making. In both the test and training set confusion matrices, we had 0 false negatives! We also had 2 false positives in the test set and 2 false postives in the training set. 

Sensitivity is the true positive rate. It is calculated as True Positives / (True Positive + False Negatives). In other words, out of all our positive cases, what proportion does the model actually correctly label? The sensitivity of this model was 1.0, which means that it is able to correctly identify 100% of adults who had CKD. This makes sense since there were 0 false negatives.

Specificity is the false positive rate. It is calculated as False Positives / (False Positives + True Negatives). In other words, out of all the negative cases we had, how many did it label as positive? Our model's specificity was 0.60. This is OK, but it means the model is labelling a lot of negative cases as positive. This makes sense since we can see that the model has those false positives in the test and training sets as mentioned before.

Positive predictive value is the proportion of predicted positive cases that are truly positive.
It is calculated as True Positives / (True Positives + False Positives). In other words, if you were diagnosed by this algorithm as having CKD, what is the likelihood of you really having CKD. In this case it is 90%, which could be nice as you would have a 10% chance of not having to deal with CKD even if you were flagged as having it.

Negative predictive value is the proportion of negative cases that are truly negative.
It is calculated as False Negatives / (False Negatives + True Negatives). In other words, if you were diagnosed as not having CKD by this algorithm, what is the likelihood that you really don't have CKD? In this case is 100%! Which makes sense since there were no false negatives.
