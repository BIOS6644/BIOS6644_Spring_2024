```python
# Mini wrangle goals:
#  1) remove stray space characters
#  2) Correct "brother" to be "father"
#  3) Write as 4 sentences, returning the "." characters

for chunk in string2split.split("."):
    print(chunk)
    chunk=chunk.strip()
    print(chunk) # strip removes whitespace from the ends

    chunk = chunk.replace('brother', 'father')
    print(chunk+'.')
    
```

```python
#My code
#This is an explanation of the above code step by step

for chunk in string2split.split("."):
    print("This is the original line without the period, but still with the spaces: ", chunk)
    #Now we will remove the spaces at the start and ends
    chunk = chunk.strip()
    print("Now we have removed the spaces: ", chunk)
    #Now we will have it seach for the worder brother and replace it with father
    chunk = chunk.replace("brother", "father")
    print(chunk, ".", sep = "")
    
```

```python
for chunk in string2split.split("."):
    chunk=chunk.strip()

    chunk = chunk.replace('brother', 'father')
    print(chunk+'.')
```

```python
#This is the above code all done at once, without the extra print statements
#This is how I would do this on my own

for chunk in string2split.split("."):
    chunk = chunk.strip()
    chunk = chunk.replace("brother", "father")
    print(chunk, ".", sep = "")
```

```python
#A different approach using a for-loop shortcut called
#    a List Comprehension
output = [xx.strip().replace("brother", "father") for xx in string2split.split(".")]
print(output)

print('.\n'.join(output))
```

```python
#My code
#here we're doing the same thing but with a for loop
#This is with a "list comprehension"
#It's the fancy one line way to print a for loop
#and it's faster for huge data sets
#Thing you're doing at the front, for loop at the end
output = [chunk.strip().replace("brother", "father") for chunk in string2split.split(".")]

print(".\n".join(output)
```

```python
#This is obtuse
#even for people who understand it
#He wouldn't even do this for anything that he expects him or anyone else to ever come back to

# as a 1-liner
print('.\n'.join([xx.strip().replace("brother", "father") for xx in string2split.split(".")]))
```
