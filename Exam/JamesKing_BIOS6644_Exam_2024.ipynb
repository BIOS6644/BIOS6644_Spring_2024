```python
arms_df.drop('Total world import', inplace=True)
max_all=arms_df['2023'].argmax()
max_country = arms_df.index[max_all]
#arms_df.iloc[max_country]

#
print(f'Country with most arms imports in 2023: {max_country}')
#
```

```python

mdf = arms_df.join(pop_df)
mdf['percap'] = mdf['2023'] / mdf.population *1e6
mdf.sort_values('percap',ascending=False, inplace=True)
mdf
print(f'Country with highest per-capita arms imports: {mdf.iloc[0].name}')
```

```python
mdf[0:10].plot.barh(y='percap')
import pylab as plt
plt.title('Top Weapons Importers, 2023')
plt.xlabel('Per Capita TIV\nhttps://www.sipri.org/')
```

According to the Stockholm International Peace Research Institute (SIPRI), the country with the the most 
arms imports in 2023 is (unsurprisingly) Ukraine.

SIPRI uses an index to rate the "value" of the arms transfers which is designed to remove the effects of 
inflation, currency exchange rates, etc.   It accounts only for "major" weapons sysems such as air defence, tanks,
etc., and does not attempt to count small arms nor their ammunition.
