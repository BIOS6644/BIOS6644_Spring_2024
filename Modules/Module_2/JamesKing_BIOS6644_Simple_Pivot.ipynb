```python
# The problem as you described it would have a single data frame 
# and that data frame would have a subject column which matches the key for the 
# dictionary
#
# One way to do this would be to loop through the df_wb dictionary and add the column
# you want, then at the end, merge them all together

allsheets = []
for this_sheet in df_wb.items():
    this_subj_id = this_sheet[0]
    this_subj_data = this_sheet[1]
    
    this_subj_data['SubjID'] = this_subj_id # Make a new column with every entry as the subj ID
    allsheets.append(this_subj_data)
    
```

```python
print(allsheets) # a list with a separate dataframe for each patient
```

```python
df = pd.concat(allsheets)
df
```

```python
# A slightly idfferent approach:
# Read it in as already-pivoted
```

```python
df_wb = pd.read_excel(f,index_col=0,sheet_name=None)
allsheets=[]

for this_sheet in df_wb.items():
    this_subj_id = this_sheet[0]
    this_subj_data = this_sheet[1]
    
    this_subj_data = this_subj_data.T  # transpose
    this_subj_data.index = pd.Index([this_subj_id])
    allsheets.append(this_subj_data)
    
```

```python
df = pd.concat(allsheets)
df
```
