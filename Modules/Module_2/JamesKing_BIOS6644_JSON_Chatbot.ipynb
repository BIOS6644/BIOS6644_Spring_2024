# Seek questions using interrogatives

```python
# make a column with all responses as single strings (they import as a list)
df['RespStrings'] = [xx.replace('"','').replace("'","").strip(']').strip('[')  for xx in df.responses.astype(str)]
df.RespStrings
```

```python
# Double-loop method

interrogs = ["what", "which", "when", "where", "who", "whom", "whose", "why","how","have"]
all_is_interrog=[]
for thisresp in df.RespStrings:
    isintg=False
    for thisintg in interrogs:
        if thisintg in thisresp.lower():
            isintg = True
    all_is_interrog.append(isintg)
all_is_interrog
```

```python
# List comprehension way
df['IsInterrog']=[max([ig in xx.lower() for ig in interrogs]) for xx in df.RespStrings]

```

```python
all(all_is_interrog==df.IsInterrog)
```
