# Regular Expressions

Regular expressions (regex--often pronounced "rej ex") are a mechanism for searching text to find arbitrarily complex patterns.

Common examples:

- Find blocks of code which match the email address pattern:  something @ somewhere . extension
- Search for phone numbers
- Find inline citations

There are various "implementations" of the idea of regexes.  Unix/Linux and Mac all have the idea baked into some command line commands.  All major programming languages implement them in some fashion, though there are ALWAYS subtle differences which are really, really annoying.

Here we'll discuss the basics built-in Python implementation.


## Python RegEx

1. Define a "pattern" by mixing letters, numbers, special-meaning characters, etc.
 - thing-numberofthing thing-numberofthing thing-numberofthing thing-numberofthing
 - "thing" can be defined as a "notthing"
 - can be combined with and/or/not logic
2. "Compile" the pattern
3. Apply the pattern to a block of text
4. Examine the results

```python
help(re)  # help files are your friend
```

```python
TextBlock="""
Dear Diary,

My teachers are such jerks.  The math guy, Mr. King, is really a complete tool!  
He's always trying to tell me how I'll use math in real life and I'm like yawhateverdude.  Blah-
blah-bla^%$^$

I wish he'd just shut up
already.  

Oh, and he looks like Mr.Potato Head.

Mrs. Dorking (how's THAT for a name!) teaches metal shop.  She's way cooler.

Metallica rules!

PS: I hate Ms. Blakely and Miss St. John, too.

Missing
"""
```

```python
re1 = r' .*[Kk]ing'
theregex = re.compile(re1)
```

```python
#re.findall?
```

```python
re.findall(theregex, TextBlock)
```

```python
# Search for Teacher names
# Desired Results:
# Mr. King
# Mr.Potato Head
# Mrs. Dorking
# Ms. Blakely
# Miss St. John
# Also, find Mx. Someonesname as well

re2 = r'Mr.'   #---That dot is a special charactere.  "Any character"
theregex = re.compile(re2)
re.findall(theregex, TextBlock)
```

```python
re2 = r'Mr\.'   #---That dot is a special charactere.  use an escape to say you literally want a dot
theregex = re.compile(re2)
re.findall(theregex, TextBlock)
```

```python
re2 = r'M[risx][s\.][s\.]? ?[A-Z][a-z]*'
theregex = re.compile(re2)
re.findall(theregex, TextBlock)
```

```python
#  Search for Mr. then a space then a word starting with a capital letter

re2 = r'Mr\. [A-Z]\D* '   #---That dot is a special charactere.  "Any character"
theregex = re.compile(re2)
re.findall(theregex, TextBlock)
```

```python
# This should catch Mr. Potato, though, right?
re2 = r'Mr\. [A-Z]*'   #---That dot is a special charactere.  "Any character"
theregex = re.compile(re2)
re.findall(theregex, TextBlock)
```
