```python
# Find patients noted as being in their 50s in the transcription
# 53-year-old.   Age 5*.  Aged 5. ...

# make a single text block for easier trial and error.

tt = "\n".join([str(cc) for cc in df.transcription])

oth = r".{0,20}5\d.{0,20}"

re.findall(oth, tt)
```

```python
# Clearly 5* doesn't work well.
# Here I see 51-year-old and 51 year old close together.   They might also use "51 yo" or "51yo"

oth = r".{0,10} 5\d.? y.{0,20}"

re.findall(oth, tt)
```

```python
# Ok--now I see "y/o" and "yo" also  'passed away 50 years ago and two bro',
# Major patterns  "y/o", "yo", "year old", "-year-old"
# Anti patterns "years old" is often referring to history
p1 = r".{0,20}5\d[-, ]yo.{0,20}"
p2 = r".{0,20}5\d[-, ]y/o.{0,20}"
p3 = r".{0,20}5\d year old .{0,20}"
p4 = r".{0,20}5\d-year-old .{0,20}"
oth = p1 + '|'+ p2 + '|' + p3 + '|' + p4
#oth = p3+'|'+p4
#oth= p4
print(oth)
re.findall(oth, tt)
```

```python
help(re)
```

```python
# Lets try to make a label column=

# loop through each row.  Check for match row-wise

oth_hits = []
for thisrow in df.itertuples():
    hitcount=re.findall(oth, str(thisrow.transcription))
    if len(hitcount)>0:
        oth_hits.append(True)
    else:
        oth_hits.append(False)
```

```python
df['InFifties'] = oth_hits
df 
```

```python
pprint.pp(df.loc[df.InFifties].sample().iloc[0]['transcription'])
```
