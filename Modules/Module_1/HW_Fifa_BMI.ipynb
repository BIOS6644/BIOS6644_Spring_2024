```python
# Strings can be "split" into a list, on a boundary marker which you provide
string2split = '''Hello.  My name is Inigo Montoya. You killed my brother.  Prepare to die'''
print(string2split)

# split on "." 
# Note that a list of strings is returned.
# and that the split boundary character is removed.
string2split.split(".")

```

```python
# Mini wrangle goals:
#  1) remove stray space characters
#  2) Correct "brother" to be "father"
#  3) Write as 4 sentences, returning the "." characters

for chunk in string2split.split("."):
    print(chunk)
    chunk=chunk.strip()
    print(chunk) # strip removes whitespace from the ends

    chunk = chunk.replace('brother', 'father')
    print(chunk+'.')
    
```

```python
for chunk in string2split.split("."):
    chunk=chunk.strip()

    chunk = chunk.replace('brother', 'father')
    print(chunk+'.')
```

```python
#A different approach using a for-loop shortcut called
#    a List Comprehension
output = [xx.strip().replace("brother", "father") for xx in string2split.split(".")]
print(output)

print('.\n'.join(output))
```

```python
# as a 1-liner
print('.\n'.join([xx.strip().replace("brother", "father") for xx in string2split.split(".")]))
```

```python
# file path
file_path = r"C:\education\Spring2024\BIOS 6644\Data\fifa_players.csv"


def calculate_bmi(height, weight):
    height = float(height)/100
    weight = float(weight)
    bmi = weight/(height**2)
    return str(bmi)

# opening file and read files
list_lines = []
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()
    for line in lines[1:]:
        line = line.strip().split(',')
        list_lines.append(''.join([line[1], ',', line[4], ',', line[5], ',', calculate_bmi(line[4], line[5])]))

# write into csv
subset_file_path = r"C:\education\Spring2024\BIOS 6644\Data\fifa_players_subset.csv"
with open(subset_file_path, 'w', encoding='utf-8') as csv_file:
    list_lines = ['full_name,height_cm,weight_kg,bmi'] + list_lines
    for line in list_lines:
        csv_file.write(line + '\n')
    
```
