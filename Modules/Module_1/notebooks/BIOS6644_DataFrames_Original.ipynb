```python
# Goal: Analyze Players' info in subsets based on nationality
#  
#
# General info & data here:
#   https://www.kaggle.com/datasets/maso0dahmed/football-players-data
#
# I unzipped mine put it in the data folder
data = '../data/fifa_players.csv'
```

```python
# Pandas has VERY strong utilities for easily importing different data types
#
#  The most common is some kind of "tabular" data--int this case, a csv
```

```python
import pandas as pd
#See options:
pd.read_csv?
```

```python
df = pd.read_csv(data)
df
```

```python
# See all columns
print(f"Column Names:  {df.columns}")
```

```python
# Get  summaries of numerical fields
df.describe()
```

```python
# Slice out columns
df[['full_name','nationality', 'age','wage_euro']]
```

```python
# Give a subset a convenience name:
cdf = df[['full_name','nationality', 'age','wage_euro']]
```

```python
cdf
```

```python
#Sort it!
cdf.sort_values('nationality', inplace=True)
```

```python
cdf
```

```python
# Getting subset of rows

cdf.loc[cdf['nationality']=='United States']

# .loc  ->  Select by True/False
```

```python
cdf['nationality']=='United States'
```

```python
cdf.loc[cdf['nationality'].str.startswith('United')]
```

```python
countries = cdf.nationality.unique()
countries
```

```python
check_country = ['Belize', 'United Kingdom', 'Malaysia', 'Singapore', 'Kyrgyzstan', 
                 'Tajikistan', 'Mongolia', 'Mauritius', 'Nepal', 'Bhutan', 'Taiwan']
for cc in check_country:
    print(f'{cc} in list: {cc in countries}')
```

```python
stans = cdf.nationality.str.endswith('stan')
cdf.loc[stans].nationality.unique()
```
