```python
# Goal: Analyze Players' info in subsets based on nationality
#  
#
# General info & data here:
#   https://www.kaggle.com/datasets/maso0dahmed/football-players-data
#
# I unzipped mine put it in the data folder
data = '../data/fifa_players.csv'
```

```python
import pandas as pd
```

```python
df = pd.read_csv(data)
df
```

```python
# See all columns
print(f"Column Names:  {df.columns}")
```

```python
# Get  summaries of numerical fields
df.describe()
```

```python
import numpy as np
```

```python
# loc is "label based"  iloc is "location based"

dfi = df.copy()
dfi.index = pd.Index(np.array(dfi['age']))  #np.array strips of pandas-specific stuff like the column name
dfi
```

```python
dfi.iloc[42]
```

```python
dfi.loc[42]
```

```python
dfi.boxplot(column='wage_euro', by='age')
```

```python
import pylab as plt
plt.figure()
plt.plot(df.age, df.wage_euro)
```

```python
plt.figure()
plt.plot(df.age, df.wage_euro, 'r.', markersize=1, alpha=0.2)
plt.title('Footballer Wage by Age')
plt.xlabel('Age (years)')
plt.ylabel(r'Wage (Euros)')
```
