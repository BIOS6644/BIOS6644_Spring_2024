```python
# The .str is a "indexer method" which allows you to select stuff based
# on the resuts of string methds.

# Some handy string methods:
#   s.startswith('thing')   checks to see if string s begins with "thing"
#   s.endswith('thing')     checks to see if string s ends with 'thing'
#   s.upper()               returns the string s coverted to UPPER CASE
#   s.lower()               returns s converted to lower case
#   s.capitalize()          returns s with Capital first letter


svar = "I like TRAFFIC LIGHTS."

print(f'.startswith("I"):       {svar.startswith("I")}')
print(f'.endswith("I"):         {svar.endswith("I")}')
print(f'.upper():               {svar.upper()}')
print(f'.lower():               {svar.lower()}')
print(f'.capitalize():          {svar.capitalize()}')
print('\n')
print(f'.find("LI"):            {svar.find("LI")}')
print(f'.upper().find("LI"):    {svar.upper().find("LI")}')


#cdf.loc[cdf['nationality'].str.startswith('United')]
```

```python
# Select rows where nationality ends with 'stan'
stans = cdf.nationality.str.endswith('stan')
cdf.loc[stans].nationality.unique()

# Pakistan?  Kyrgyztan?  Tajikistan?  
```

```python
# Select rows where nationality ends with 'land'
lands = cdf.nationality.str.endswith('land')
cdf.loc[lands].nationality.unique()

# Greenland?  Others?
```

```python
# Countries with 2+ words
hasspace = cdf.nationality.str.find(' ')>=0
cdf.loc[hasspace].nationality.unique()
```
