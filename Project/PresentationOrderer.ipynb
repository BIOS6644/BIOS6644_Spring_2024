```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
```

```python

dt1 = dt(year=2024, month=5, day=6)
dt2 = dt(year=2024, month=5, day=8)
dt3 = dt(year=2024, month=5, day=13)
dt4 = dt(year=2024, month=5, day=15)

session_dates = [dt1]*5+[dt2]*5+[dt3]*5+[dt4]*4
session_dates
session_slots = np.array(list(range(5))*4)+1
```

```python
# Let session number be 5/6: 1-5, 5/8: 6-10, ...
sdf = pd.read_csv('students.txt', header=None)
sdf.columns = ['Student']
rng = np.random.RandomState(42)
sdf['rand_number']=rng.rand(len(sdf))
sdf.sort_values('rand_number', inplace=True)
#sdf.reset_index(inplace=True,drop=True)
sdf['presentation_date'] = session_dates
sdf['presentation_slot'] = session_slots[0:len(sdf)]
sdf
```

```python
a_sdf = sdf.copy()
a_sdf.sort_index(inplace=True)
a_sdf
```
