```python
##################################################################################################################
#
# YouDo:
# 
#    1) Get rid of any columns that have only NaN values
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html
#
#######################################  BEGIN STUDENT CODE  #####################################################

df = df.dropna(axis='columns', how='all')
print(df)

#######################################   END STUDENT CODE   #####################################################

assert df.shape == (92, 125)
```

```python
##################################################################################################################
#
# YouDo:
# 
#    1) Generate boxplots of Tryptophan levels separated based on 
#        the "Label" value
#
#      Use whatever plotting library you like.   
#      Hint 1: matplotlib has a boxplot tool
#          help(plt.boxplot)
#      Hint 2: pandas data frames have a built-in boxplot tool
#          help(df.boxplot)
#
#######################################  BEGIN STUDENT CODE  #####################################################

df.boxplot(column='Tryptophan', by='Label', grid=True)

plt.title('Tryptophan')
plt.xlabel('Label Value')
plt.ylabel('Tryptophan Levels')

plt.show()

#######################################   END STUDENT CODE   #####################################################
```

```python
##################################################################################################################
#
# YouDo:
# Make a pair of plots as follows:
#    1) Show the correlation matrix between columns of the data
#        (hint 1: it should be 124x124, hint 2: check out plt.imshow() 
#         Hint 2: numpy has a function for generating the correlation matrix)
#
#    2) Make a similar plot showing correlation > 90%
#
#  Stretch goal:  using plt.subplot()  make a single figure with both plots side-by-side
#
#######################################  BEGIN STUDENT CODE  #####################################################

df.replace(['NA'], np.nan, inplace=True)

corr_matrix = np.corrcoef(df.select_dtypes("number"), rowvar=False)

fig, ax1 = plt.subplots(1, 2, figsize=(10, 4))

im1 = ax1[0].imshow(corr_matrix, cmap='viridis', aspect='auto')
ax1[0].set_title('Correlation Matrix')
cbar1 = fig.colorbar(im1, ax=ax1[0])

corr_matrix2 = np.clip(corr_matrix, a_min=0.9, a_max=None)

im2 = ax1[1].imshow(corr_matrix2, cmap='viridis', aspect='auto')
ax1[1].set_title('Correlation Matrix II')
cbar2 = fig.colorbar(im2, ax=ax1[1])

plt.tight_layout()
plt.show()

#######################################   END STUDENT CODE   #####################################################
```

```python
##################################################################################################################
# YouDo:
#  1) Remove any rows with na values
# 
#  2) Split the data into "test" and "train" bits with ~30% in the test set.
#       The "Label" column is the "y" data.  All other columns are the "X" data
#       Give them these names:
#       X_train, X_test, y_train, y_test
#######################################  BEGIN STUDENT CODE  #####################################################

# 1)

df_cleaned = df.dropna()
#print(df_cleaned)

#row_index = 33
#print(df_cleaned.iloc[row_index])

#2)

from sklearn.model_selection import train_test_split

X = df_cleaned.iloc[:,1:] 
y = df_cleaned.iloc[:,0] 

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=104, test_size=0.3
)

print('X_train: ')
print(X_train.head())
 
print('\nX_test:')
print(X_test.head())
 
print('\ny_train:')
print(y_train.head())
 
print('\ny_test:')
print(y_test.head())

#######################################   END STUDENT CODE   #####################################################
```

```python
#Build a model to classify the Label based on the other columns

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=8675309)
clf.fit(X_train, y_train)
```

```python
# Test the model
predictions = clf.predict(X_test)
reality = y_test
```

```python
# YouDo:
#  1) Evaluate the "goodness" of the predictions and write a paragraph describing the results as a markdown.
#
#######################################  BEGIN STUDENT CODE  #####################################################

reality-np.array(predictions)

#######################################   END STUDENT CODE   #####################################################
```
