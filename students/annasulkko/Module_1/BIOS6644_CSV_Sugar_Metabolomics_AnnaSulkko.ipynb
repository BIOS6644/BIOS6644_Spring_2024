```python
##################################################################################################################
#
# YouDo:
#  1) Remove any rows with na values
# 
#  2) Split the data into "test" and "train" bits with ~30% in the test set.
#       The "Label" column is the "y" data.  All other columns are the "X" data
#       Give them these names:
#       X_train, X_test, y_train, y_test

#######################################  BEGIN STUDENT CODE  #####################################################

####################################### Task 1 :  Remove NA values from rows  #####################################################
corr_matrix_cleaned = corr_matrix.dropna(axis=0, how='all').dropna(axis=1, how='all')
corr_matrix_cleaned



####################################### Task 2 :  Split Data into "test" and "train"  #####################################
# The majority of the code below is from ChatGPT3.5 

from sklearn.model_selection import train_test_split

# Declaring 'Labels' as integers to prevent Random Forest Classifer from throwing error 
corr_matrix_cleaned['Label'] = corr_matrix_cleaned['Label'].astype(int)
print(corr_matrix_cleaned['Label'].dtype)


# Dropping 'Label' column from "X" data 
X = corr_matrix_cleaned.drop(columns=['Label'])


# Assigning 'y' data to 'Label' column
y = corr_matrix_cleaned['Label']

# Split the data into training and testing sets with 70% training and 30% testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(X_train)
print(X_test)
print(y_train)
print(y_train)




#######################################   END STUDENT CODE   #####################################################
```
