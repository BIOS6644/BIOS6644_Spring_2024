```python
##################################################################################################################
#
# YouDo:
#  1) Remove any rows with na values
# 
#  2) Split the data into "test" and "train" bits with ~30% in the test set.
#       The "Label" column is the "y" data.  All other columns are the "X" data
#       Give them these names:
#       X_train, X_test, y_train, y_test

#######################################  BEGIN STUDENT CODE  #####################################################

# 1)Remove any rows with na values
dat = df_m.dropna() # drop rows with nan values
dat

 
# 2) Split the data into "test" and "train" bits with ~30% in the test set.
#       The "Label" column is the "y" data.  All other columns are the "X" data
#       Give them these names:
#       X_train, X_test, y_train, y_test
#https://machinelearningmastery.com/train-test-split-for-evaluating-machine-learning-algorithms/
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
# split into train test sets
from sklearn.model_selection import train_test_split
xcols = dat.columns[1:]
X_train, X_test, y_train, y_test = train_test_split(dat[xcols], dat['Label'],test_size = 0.3)



#######################################   END STUDENT CODE   #####################################################
```

```python
#Build a model to classify the Label based on the other columns

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=8675309)
clf.fit(X_train, y_train)
```

```python
# Test the model
predictions = clf.predict(X_test)
reality = y_test
```

```python
# YouDo:
#  1) Evaluate the "goodness" of the predictions and write a paragraph describing the results as a markdown.
#
#######################################  BEGIN STUDENT CODE  #####################################################
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
print(classification_report(predictions,reality))

#######################################   END STUDENT CODE   #####################################################
```

# Analysis
It goes here.

```python
Precision: Out of our sample size of n=92 that the model predicted would have for our label variable 1 as response, 100% actually did.

Recall: Out of all the sample size that actually did get a 1 response for the label variable, the model only predicted this outcome correctly for 88% of those.

F1 Score: 2 * (Precision * Recall) / (Precision + Recall)= 2 (1 * 0.88) / (1 + 0.88)=0.94.

Since this value is very close to 1, it tells us that the model does a good job of predicting whether or not the population has a 0 or 1 as response.
```
