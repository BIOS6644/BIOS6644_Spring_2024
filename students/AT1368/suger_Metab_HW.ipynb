```python
import pandas as pd

# open CSV file
file_path = r"C:\education\Spring2024\BIOS 6644\Data\sugar_metabolomics (1).csv"

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path , header=0, skipfooter=1, engine='python')

# the first few rows of the DataFrame
print(df)
```

```python
#Get rid of any columns that have only NaN values
df = df.dropna(axis=1, how='all')
df
```

```python
#boxplots of Tryptophan levels separated based on 
#        the "Label" value
df.boxplot(column= 'Tryptophan', by= 'Label')
```

```python
# Corrlation matrix:
corrM = df.corr()
plt.figure()
plt.imshow(corrM)
plt.show()
corrM
```

```python
#Remove any rows with na values:
df_No_na = df.dropna()

df_No_na
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
# split data to X and Y:
X = df_No_na.drop('Label', axis=1)
y = df_No_na['Label']

#  testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Print 
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=8675309)
clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
reality = y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# Calculating evaluation metrix
mae = mean_absolute_error(y_test, predictions)
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
r2 
```
