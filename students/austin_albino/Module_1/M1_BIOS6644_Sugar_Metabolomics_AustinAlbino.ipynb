```python
##################################################################################################################
#
# YouDo:
#  1) Remove any rows with na values
# 
#  2) Split the data into "test" and "train" bits with ~30% in the test set.
#       The "Label" column is the "y" data.  All other columns are the "X" data
#       Give them these names:
#       X_train, X_test, y_train, y_test

#######################################  BEGIN STUDENT CODE  #####################################################
from sklearn.model_selection import train_test_split

# 1)
df = df.dropna(axis='rows', how='any')
 
#2)
# train, test = train_test_split(df, test_size = 0.3)
# train
# test
xcols = df.columns[1:]
X_train, X_test, y_train, y_test = train_test_split(df[xcols], df['Label'],test_size = 0.3)
#######################################   END STUDENT CODE   #####################################################
```

```python
#Build a model to classify the Label based on the other columns

from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=8675309)
clf.fit(X_train, y_train)
```

```python
# Test the model
predictions = clf.predict(X_test)
reality = y_test
```

```python
# YouDo:
#  1) Evaluate the "goodness" of the predictions and write a paragraph describing the results as a markdown.
#
#######################################  BEGIN STUDENT CODE  #####################################################
# print(predictions,reality)
type(predictions)
fiteval = reality.reset_index().drop(columns='index')
fiteval['predictions'] = pd.Series(predictions)
fiteval = fiteval.rename(columns={'Label':'reality','predictions':'prediction'})
print(fiteval.corr())
pd.crosstab(fiteval['reality'],fiteval['prediction'])
#######################################   END STUDENT CODE   #####################################################
```
