```python
import pandas as pd #Use import to bring in dataset, pandas is a library
#https://pandas.pydata.org/docs/getting_started/index.html
#See options:
pd.read_csv # add ? to help you find documentation for Pandas; read_XXX (lots of types of datafiles are included, including STATA files)
#Read_table can import files that are not comma delimited
#parse_dates will separate out dates
```

```python
df = pd.read_csv(data) #df is whatever naming convention for the datafile
df #shows you a table of the data
```

```python
# See all column names
print(f"Column Names:  {df.columns}")
```

```python
# Get summaries of numerical fields
df.describe() #Describe field is from R
```

```python
# Slice out columns, pull out columns of interest
df[['full_name','nationality', 'age','wage_euro']]
```

```python
# Give a subset a convenience name:
cdf = df[['full_name','nationality', 'age','wage_euro']]#This is a subset, df is still your dataset, so you can't alter this subset
```

```python
cdf #First column is the index column/row #
```

```python
#Sort it!
cdf.sort_values('nationality', inplace=True) #I want to sort the actual data, not just what I am looking at (save the sorted state using inplace=true)
```

```python
cdf
```

```python
# Getting subset of rows

cdf.loc[cdf['nationality']=='United States'] #two locate methods, loc and iloc. 
#If you do cdf['nationality'=='United States'], it gives you true/false for each row.
#Don't have to be =, can use > or < instead.
#Can name it, US_Players = cdf.loc[]

# .loc  ->  Select by True/False
```

```python
cdf['nationality']=='United States'
```

```python
cdf.loc[cdf['nationality'].str.startswith('United')] #Finds strings that start with United.
```

```python
countries = cdf.nationality.unique() #Unique method shows you all unique entries
countries
```

```python
check_country = ['Belize', 'United Kingdom', 'Malaysia', 'Singapore', 'Kyrgyzstan', 
                 'Tajikistan', 'Mongolia', 'Mauritius', 'Nepal', 'Bhutan', 'Taiwan']
for cc in check_country:
    print(f'{cc} in list: {cc in countries}')
```
