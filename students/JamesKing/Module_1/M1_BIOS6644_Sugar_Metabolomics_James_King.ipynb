```python
##################################################################################################################
#
# YouDo:
#  1) Remove any rows with na values
# 
#  2) Split the data into "test" and "train" bits with ~30% in the test set.
#       The "Label" column is the "y" data.  All other columns are the "X" data
#       Give them these names:
#       X_train, X_test, y_train, y_test

#######################################  BEGIN STUDENT CODE  #####################################################

# 1) Remove any rows with na values
d2 = df.dropna(axis=0)
X = d2.drop('Label', axis=1)
y = d2.Label

from sklearn.model_selection import train_test_split as tts
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.3, random_state=42)
#####################################   END STUDENT CODE   #####################################################
```

```python
print(X_train.shape, X_test.shape)
```

```python
print(y_train.shape, y_test.shape)
```

```python
#Build a model to classify the Label based on the other columns

from sklearn.ensemble import RandomForestClassifier


clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=8675309)
clf.fit(X_train, y_train) #I had to change this to y_train.values.ravel() 

```

```python
# Test the model
predictions = clf.predict(X_test)
reality = y_test

print(list(predictions))
print(list(reality))
print('-------------------------------------------------------------------------')
print(list(np.array(predictions==reality,dtype=int)))
```

In this code we will calculate the accuracy score of our model on the test and training data sets.

We will also plot the confusion matrices to take a closer look and assess false positives (top right) and false negatives (bottom left):

```python
# YouDo:
#  1) Evaluate the "goodness" of the predictions and write a paragraph describing the results as a markdown.
#
#######################################  BEGIN STUDENT CODE  #####################################################
from sklearn.metrics import accuracy_score as ACC, matthews_corrcoef as MCC
from sklearn.metrics import classification_report


#Calculating the accuracy score of the testing set
test_accuracy = ACC(reality, predictions)
mcc = MCC(predictions, reality)

print("Test accuracy:", test_accuracy)
print("MCC:", mcc)

print(classification_report(predictions, reality))


#######################################   END STUDENT CODE   #####################################################
```

```python
#################################################################################################################
# YouDo:
#  1) Evaluate the "goodness" of the predictions and write a paragraph describing the results as a markdown.
#
#######################################  BEGIN STUDENT CODE  #####################################################

# https://stackoverflow.com/questions/63526955/how-to-obtain-specificity-and-negative-predictive-value-from-scikit-learn-for-mu
from sklearn.metrics import confusion_matrix, roc_auc_score, matthews_corrcoef as MCC

def custom_classification_report(y_true, y_pred):
    tp, fn, fp, tn = confusion_matrix(y_true, y_pred).ravel()
    acc = (tp+tn)/(tp+tn+fp+fn)
    sen = (tp)/(tp+fn)
    sp = (tn)/(tn+fp)
    ppv = (tp)/(tp+fp)
    npv = (tn)/(tn+fn)
    f1 = 2*(sen*ppv)/(sen+ppv)
    fpr = (fp)/(fp+tn)
    tpr = (tp)/(tp+fn)

    print('Confusion Matrix:')
    print(confusion_matrix(predictions, reality))

    print(classification_report(reality, predictions))
    
    print(f'''
                MCC:                 {round(MCC(reality, predictions),3)}
                Accuracy:            {round(acc, 3)}
                Sensitivity/Recall:  {round(sen, 3)}
                Specificity:         {round(sp, 3)}
                PPV/Precision:       {round(ppv, 3)}
                NPV:                 {round(npv, 3)}
                F1-score:            {round(f1, 3)}
                False positive rate: {round(fpr, 3)}
                True positive rate:  {round(tpr, 3)}

                AUC-ROC:             {round(roc_auc_score(reality, predictions), 3)}
          
          ''')
    
custom_classification_report(predictions, reality)
#######################################   END STUDENT CODE   #####################################################
```

On the surface, these metrics are really good--maybe this method shows promise.   The data set is small, however, and the Fisher heuristic of >=5 elements must be in each box of the confusion matrix has not been met.   This might mean we don't have sufficient data to evaulate the goodness.
